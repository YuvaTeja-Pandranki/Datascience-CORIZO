```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://www.imdb.com/search/title/?release_date=2018-01-01,2018-12-31&sort=num_votes,desc"
```

```python
page = requests.get(url)
page
```

```python
page.content
```

```python
soup = BeautifulSoup(page.content, "html.parser")
print(soup.prettify())
```

```python
scraped_movies = soup.find_all('h3' , class_='lister-item-header')
scraped_movies
```

```python
movies = []
for movie in scraped_movies:
    movie = movie.get_text().replace('\n', "")
    movies.append(movie)
movies
```

```python
scraped_runtime = soup.find_all('span' , class_ ='runtime')
scraped_runtime
```

```python
runtimes = []
for runtime in scraped_runtime:
    runtime = runtime.get_text().replace(' min', "")
    runtimes.append(runtime)
runtimes
```

```python
data = pd.DataFrame()
data['Movies Names'] = movies
data['Runtimes'] = runtimes
data.head()
```
